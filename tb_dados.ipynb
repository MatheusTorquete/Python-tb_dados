1) Importe os dados do anexo "tb_dados.csv" e crie um data frame.

4) Selecione a segunda linha e a segunda coluna do DataFrame.

5) Encontre o filme com a avaliação mais alta

```python
filme_com_maior_avaliacao = df[df['analiacoes_qtd'] == df['analiacoes_qtd'].max()]
print("\nFilme com a avaliação mais alta:")
print(filme_com_maior_avaliacao)
```
